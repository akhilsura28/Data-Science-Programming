```python
# Importing Python Libraries and Modules Demo

# Importing an entire library/module
import math
import random

# NOTE: importing a library a second (third, fourth, etc.) time will not re-import the library. This is useful, as it 
# allows you to import a library in one cell and not have to remember that you did, and then import it again 
# in a later cell. NOTE: Though, it's considered good practice to conduct all your imports at the top of your 
# notebook (so you can see what libraries you're using at a glance), it's often easy to maintain (and read) 
# code that imports in the cell that you need the library.


# Using functions from imported libraries/modules
print("Square root of 16:", math.sqrt(16))  # Output: 4.0
print("Random number between 1 and 10:", random.randint(1, 10))  # Output: Random number
```

```python
# Importing specific functions from a library/module
from datetime import datetime
from itertools import combinations

# Using functions from imported libraries/modules
current_time = datetime.now()
print("Current time:", current_time)  # Output: Current date and time

combinations_list = list(combinations([1, 2, 3], 2))
print("Combinations:", combinations_list)  # Output: [(1, 2), (1, 3), (2, 3)]
```

```python
# Importing a library/module with an alias
import pandas as pd
import numpy as np

# Using functions from imported libraries/modules with aliases
data = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
print("DataFrame:")
print(data)

array = np.array([1, 2, 3, 4, 5])
print("Array:", array)

# Importing all functions from a library/module (not recommended)
from matplotlib import *

# Using functions from imported libraries/modules
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Plot")
plt.show()
```

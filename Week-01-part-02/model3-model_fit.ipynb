# Predictive Modeling Example

## Step 1: Import the libraries we will use in this notebook

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

np.random.seed(1) # set this to ensure the results are repeatable. 
```

## Step 2: Load the given data that we will model

```python
df = pd.read_csv('../data/model3_dataset.csv')

df.head(3)
```

Seperate out the input m(X) and the target (y)

```python
X=df[['input']]
y=df[['target']]
```

## Step 3: Explore the given data

To use linear regression techniques effectively, we need to gain a thorough understanding of the data and identify potential relationships between variables. Linear regression assumes a linear relationship between the independent variable(s) and the dependent variable (note: more specifically, linear in the coefficients -- we can apply any exotic transformations to the input variable(s) as we want -- and therefore, use linear regression to fit very non-linear relationships).

However, it is important to acknowledge that linear regression can be limited by its inherent bias. Therefore, a critical step in utilizing linear regression is to explore the data and develop a theory about the relationship. This theory should be expressed as a function, representing the expected relationship between the variables.

By finding the best parameters for this function using the ordinary least squares method, we aim to minimize errors and achieve the most accurate linear regression model possible. It is through this rigorous exploration and hypothesis formulation that we can maximize the effectiveness of linear regression as a modeling technique."

This updated explanation emphasizes the importance of exploring the data, formulating a hypothesis, and finding the best parameters to minimize errors in the linear regression model. It also acknowledges the inherent bias of linear regression and the need for careful theory development.

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.scatter(X, y, color='blue')

ax.set_xlabel('input')
ax.set_ylabel('target')
plt.tight_layout()
plt.show()
```

Based on our exploration of the data, we can formulate a hypothesis about the relationship between the variables. In this case, we hypothesize that there is a linear relationship between the sine of the variable and the target variable. We can express this relationship as a function:

y = b0 + b1 * sin(x)

## Step 4: Fit the model

Fit the linear regression model to the dataset

```python
lin_reg = LinearRegression().fit(np.sin(X),y) # fit the model by first transorming the input data using our theory that the relationship involeves a sine function
```

## Step 5: Analyze the linear regression model's performance

Visualize the linear regression model results

```python
fig = plt.figure() # create a figure
ax = fig.add_subplot() # add a subplot

ax.scatter(X, y, color='blue') # plot the actual values
ax.scatter(X,lin_reg.predict(np.sin(X)),color='red') # plot the predicted values

ax.set_title('Linear Regression') 
ax.set_xlabel('input') 
ax.set_ylabel('target')
plt.tight_layout() # make sure the labels are not cut off
plt.show() # show the plot
```

```python
b0 = lin_reg.intercept_[0] # get the intercept
b1 = lin_reg.coef_[0][0] # get the slope
r2 = lin_reg.score(X, y) # get the R^2
print(f"Y = {b0:.2f} + {b1:.2f}sin(x)")
print(f"R^2: {lin_reg.score(np.sin(X), y):.3f}") 
```

## Conclusion:

The hidden model is:
* y = 20 + 30.5 * np.sin(X)

And our model is:
* y = 20.14 + 30.57 * np.sin(X)

The model is very close to the hidden model! This is strong indication that we have found the hidden relationship.
